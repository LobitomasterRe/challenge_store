### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
facturacion_tienda1 = tienda1['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

print(f"Facturación tienda 1: {facturacion_tienda1}")
print(f"Facturación tienda 2: {facturacion_tienda2}")
print(f"Facturación tienda 3: {facturacion_tienda3}")
print(f"Facturación tienda 4: {facturacion_tienda4}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

facturacion_tienda1 = tienda1['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

facturacion_tiendas = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Facturación': [facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4]
}

df_facturacion = pd.DataFrame(facturacion_tiendas)

plt.figure(figsize=(12, 8))
ax = plt.bar(df_facturacion['Tienda'], df_facturacion['Facturación'])
plt.title('Facturación por tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturación total')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)
plt.bar_label(ax, labels=[f'{x:,.0f}' for x in ax.datavalues], padding=3)
plt.show()
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda1.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría en la tienda 1:")
print(ventas_por_categoria_tienda1)
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría en la tienda 2:")
print(ventas_por_categoria_tienda2)
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría en la tienda 3:")
print(ventas_por_categoria_tienda3)
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría en la tienda 4:")
print(ventas_por_categoria_tienda4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ventas_por_categoria_tienda1 = tienda1.groupby('Categoría del Producto')['Precio'].sum()
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()

categorias = ventas_por_categoria_tienda1.index

ventas_tiendas = [ventas_por_categoria_tienda1.values,
                  ventas_por_categoria_tienda2.values,
                  ventas_por_categoria_tienda3.values,
                  ventas_por_categoria_tienda4.values]

ancho_barra = 0.2

x = np.arange(len(categorias))

fig, ax = plt.subplots(figsize=(12, 8))

ax.bar(x - ancho_barra * 1.5, ventas_tiendas[0], width=ancho_barra, label='Tienda 1')
ax.bar(x - ancho_barra * 0.5, ventas_tiendas[1], width=ancho_barra, label='Tienda 2')
ax.bar(x + ancho_barra * 0.5, ventas_tiendas[2], width=ancho_barra, label='Tienda 3')
ax.bar(x + ancho_barra * 1.5, ventas_tiendas[3], width=ancho_barra, label='Tienda 4')

ax.set_title('Ventas por categoría y tienda')
ax.set_xlabel('Categoría del Producto')
ax.set_ylabel('Ventas totales')
ax.set_xticks(x)
ax.set_xticklabels(categorias, rotation=45, ha='right')
ax.legend()
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda1['Calificación'].mean()
print(f"Calificación promedio de la tienda 1: {calificacion_promedio_tienda1}")
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
print(f"Calificación promedio de la tienda 1: {calificacion_promedio_tienda2}")
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
print(f"Calificación promedio de la tienda 1: {calificacion_promedio_tienda3}")
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
print(f"Calificación promedio de la tienda 1: {calificacion_promedio_tienda4}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

calificacion_promedio_tienda1 = tienda1['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

data_calificaciones = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]
}
df_calificaciones = pd.DataFrame(data_calificaciones)

plt.figure(figsize=(10, 6))
plt.plot(df_calificaciones['Tienda'], df_calificaciones['Calificación Promedio'], marker='o', linestyle='-', color='blue')
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)  # Agregar una cuadrícula
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
productos_mas_vendidos_tienda1 = tienda1['Producto'].value_counts().sort_values(ascending=False)
productos_menos_vendidos_tienda1 = tienda1['Producto'].value_counts().sort_values(ascending=True)

print("Productos más vendidos en la tienda 1:")
print(productos_mas_vendidos_tienda1.head())

print("Productos menos vendidos en la tienda 1:")
print(productos_menos_vendidos_tienda1.head())

productos_mas_vendidos_tienda2 = tienda2['Producto'].value_counts().sort_values(ascending=False)
productos_menos_vendidos_tienda2 = tienda2['Producto'].value_counts().sort_values(ascending=True)

print("Productos más vendidos en la tienda 2:")
print(productos_mas_vendidos_tienda2.head())

print("Productos menos vendidos en la tienda 2:")
print(productos_menos_vendidos_tienda2.head())

productos_mas_vendidos_tienda3 = tienda3['Producto'].value_counts().sort_values(ascending=False)
productos_menos_vendidos_tienda3 = tienda3['Producto'].value_counts().sort_values(ascending=True)

print("Productos más vendidos en la tienda 3:")
print(productos_mas_vendidos_tienda3.head())

print("Productos menos vendidos en la tienda 3:")
print(productos_menos_vendidos_tienda3.head())

productos_mas_vendidos_tienda4 = tienda4['Producto'].value_counts().sort_values(ascending=False)
productos_menos_vendidos_tienda4 = tienda4['Producto'].value_counts().sort_values(ascending=True)

print("Productos más vendidos en la tienda 4:")
print(productos_mas_vendidos_tienda4.head())

print("Productos menos vendidos en la tienda 4:")
print(productos_menos_vendidos_tienda4.head())


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


productos_mas_vendidos_tienda1 = tienda1['Producto'].value_counts().sort_values(ascending=False)
productos_menos_vendidos_tienda1 = tienda1['Producto'].value_counts().sort_values(ascending=True)

productos_mas_vendidos_tienda2 = tienda2['Producto'].value_counts().sort_values(ascending=False)
productos_menos_vendidos_tienda2 = tienda2['Producto'].value_counts().sort_values(ascending=True)

productos_mas_vendidos_tienda3 = tienda3['Producto'].value_counts().sort_values(ascending=False)
productos_menos_vendidos_tienda3 = tienda3['Producto'].value_counts().sort_values(ascending=True)

productos_mas_vendidos_tienda4 = tienda4['Producto'].value_counts().sort_values(ascending=False)
productos_menos_vendidos_tienda4 = tienda4['Producto'].value_counts().sort_values(ascending=True)

top_5_productos = pd.concat([productos_mas_vendidos_tienda1.head(5),
                              productos_mas_vendidos_tienda2.head(5),
                              productos_mas_vendidos_tienda3.head(5),
                              productos_mas_vendidos_tienda4.head(5)]).index.unique()
df_top_5 = pd.DataFrame({
    'Tienda 1': productos_mas_vendidos_tienda1.loc[top_5_productos],
    'Tienda 2': productos_mas_vendidos_tienda2.loc[top_5_productos],
    'Tienda 3': productos_mas_vendidos_tienda3.loc[top_5_productos],
    'Tienda 4': productos_mas_vendidos_tienda4.loc[top_5_productos]
}).fillna(0)

df_top_5.plot(kind='bar', figsize=(12, 8), width=0.8)
plt.title('Top 5 Productos Más Vendidos por Tienda')
plt.xlabel('Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


productos_menos_vendidos_tienda1 = tienda1['Producto'].value_counts().sort_values(ascending=True)
productos_menos_vendidos_tienda2 = tienda2['Producto'].value_counts().sort_values(ascending=True)
productos_menos_vendidos_tienda3 = tienda3['Producto'].value_counts().sort_values(ascending=True)
productos_menos_vendidos_tienda4 = tienda4['Producto'].value_counts().sort_values(ascending=True)

top_5_productos = pd.concat([productos_menos_vendidos_tienda1.head(5),
                              productos_menos_vendidos_tienda2.head(5),
                              productos_menos_vendidos_tienda3.head(5),
                              productos_menos_vendidos_tienda4.head(5)]).index.unique()
df_top_5 = pd.DataFrame({
    'Tienda 1': productos_menos_vendidos_tienda1.loc[top_5_productos],
    'Tienda 2': productos_menos_vendidos_tienda2.loc[top_5_productos],
    'Tienda 3': productos_menos_vendidos_tienda3.loc[top_5_productos],
    'Tienda 4': productos_menos_vendidos_tienda4.loc[top_5_productos]
}).fillna(0)

df_top_5.plot(kind='bar', figsize=(12, 8), width=0.8)
plt.title('Top 5 Productos Menos Vendidos por Tienda')
plt.xlabel('Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

envio_promedio_tienda1 = tienda1.groupby('Lugar de Compra')['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2.groupby('Lugar de Compra')['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3.groupby('Lugar de Compra')['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4.groupby('Lugar de Compra')['Costo de envío'].mean()

df_envio = pd.DataFrame({
    'Tienda 1': envio_promedio_tienda1,
    'Tienda 2': envio_promedio_tienda2,
    'Tienda 3': envio_promedio_tienda3,
    'Tienda 4': envio_promedio_tienda4
}).fillna(0)

plt.figure(figsize=(12, 8))
sns.heatmap(df_envio, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Costo Promedio de Envío por Tienda y Lugar de Compra')
plt.xlabel('Tienda')
plt.ylabel('Lugar de Compra')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

envio_promedio_tienda1 = tienda1.groupby('Lugar de Compra')['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2.groupby('Lugar de Compra')['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3.groupby('Lugar de Compra')['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4.groupby('Lugar de Compra')['Costo de envío'].mean()

df_envio = pd.DataFrame({
    'Tienda': ['Tienda 1'] * len(envio_promedio_tienda1) + ['Tienda 2'] * len(envio_promedio_tienda2) + ['Tienda 3'] * len(envio_promedio_tienda3) + ['Tienda 4'] * len(envio_promedio_tienda4),
    'Lugar de Compra': list(envio_promedio_tienda1.index) + list(envio_promedio_tienda2.index) + list(envio_promedio_tienda3.index) + list(envio_promedio_tienda4.index),
    'Costo de envío': list(envio_promedio_tienda1.values) + list(envio_promedio_tienda2.values) + list(envio_promedio_tienda3.values) + list(envio_promedio_tienda4.values)
})

df_envio = df_envio.sort_values('Costo de envío', ascending=False)

plt.figure(figsize=(12, 8))
plt.bar(df_envio['Lugar de Compra'], df_envio['Costo de envío'], color='skyblue')
plt.title('Costo Promedio de Envío por Lugar de Compra')
plt.xlabel('Lugar de Compra')
plt.ylabel('Costo de envío')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

**Informe Final: Recomendación de Venta de Tienda para el Sr. Juan**

**Introducción:**

El presente informe tiene como objetivo analizar el rendimiento de cuatro tiendas (Tienda 1, Tienda 2, Tienda 3 y Tienda 4) propiedad del Sr. Juan, con el fin de recomendar cuál de ellas debería considerar vender. El análisis se basa en datos de ventas, incluyendo la facturación total, las ventas por categoría de producto, la calificación promedio de la tienda, los productos más y menos vendidos, y el costo promedio de envío. El objetivo es identificar la tienda con menor rendimiento y potencial de crecimiento, para que el Sr. Juan pueda tomar una decisión informada y estratégica.

**Desarrollo:**

**1. Análisis de Facturación:**

1.png

La Tienda 1 presenta la facturación total más alta, seguida por la Tienda 2, Tienda 3 y finalmente la Tienda 4. Esto indica que la Tienda 1 es la que genera más ingresos y tiene una mayor participación en el mercado. La Tienda 4, por otro lado, presenta la facturación total más baja, lo que sugiere un menor rendimiento en términos de ingresos.

**2. Ventas por Categoría:**

2.png

El gráfico de ventas por categoría muestra que las tiendas tienen un rendimiento similar en la mayoría de las categorías. Sin embargo, se pueden observar algunas diferencias. Por ejemplo, la Tienda 1 tiene un mayor volumen de ventas en la categoría "Electrónicos", mientras que la Tienda 3 destaca en la categoría "Muebles". La Tienda 4 muestra un menor volumen de ventas en "Electrodomésticos" en comparación con las otras tiendas.

**3. Calificación Promedio de la Tienda:**

3.png

La Tienda 3 tiene la calificación promedio más alta, seguida por la Tienda 2, Tienda 4 y finalmente la Tienda 1. Esto indica que la Tienda 3 tiene la mayor satisfacción del cliente, mientras que la Tienda 1 podría tener áreas de mejora en la experiencia del cliente.

**4. Productos Más y Menos Vendidos:**

4.png

5.png

Los gráficos de productos más y menos vendidos muestran las diferencias en las preferencias de los clientes en cada tienda. La Tienda 1, por ejemplo, tiene como productos más vendidos "Microondas", "TV LED UHD 4K" y "Armario", mientras que la Tienda 4 destaca en la venta de "Cama box", "Cubertería" y "Dashboards con Power BI". Es importante analizar estos patrones para comprender las fortalezas y debilidades de cada tienda en términos de oferta de productos.

**5. Costo Promedio de Envío:**

6.png

El análisis del costo promedio de envío muestra que existen variaciones en los costos dependiendo de la tienda y del lugar de compra. La Tienda 4 podría tener costos de envío ligeramente más altos en promedio, especialmente en ciertas zonas. Esto podría afectar su rentabilidad y competitividad en comparación con las demás tiendas.

Con base en el análisis realizado, se recomienda al Sr. Juan considerar la venta de la **Tienda 4**.

La **Tienda 4** presenta un menor rendimiento general en comparación con las demás tiendas en varios indicadores clave:

**Facturación total:**

Tiene la facturación total más baja, lo que indica un menor rendimiento en términos de ingresos.

**Calificación promedio:**

Su calificación promedio, aunque no es la más baja, está por debajo de la Tienda 2 y la Tienda 3, sugiriendo una menor satisfacción del cliente.

**Ventas por categoría:**

**Costo promedio de envío:**

Podría tener costos de envío ligeramente más altos en promedio, lo que afectaría su rentabilidad.

Si bien la *Tienda 4* podría tener algunas fortalezas, como la venta de ciertos productos específicos, su bajo rendimiento en facturación total y otros indicadores clave la convierten en la candidata más adecuada para la venta.
